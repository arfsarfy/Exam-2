### 5. Make Table like this to describe the dataframe

#### Columns Explanations :
 - __dataFeatures__ = column name
 - __dataType__ = data type of that column
 - __null__ = how many null data in that column/series
 - __nullPct__ = null percentage of total data in that column
 - __unique__ = how many unique data
 - __uniqueSample__ = get some unique sample from each column

#### hint : search .sample(), .drop_duplicates(), .values, .isna(), dtype

```python
data=[]
for item in dfTitanic.columns:
    data.append([
                item,
                dfTitanic.dtypes.loc[item],
                dfTitanic[item].isna().sum(),
                dfTitanic[item].isna().sum()/len(dfTitanic[item])*100,
                dfTitanic[item].nunique(),
                list(dfTitanic[dfTitanic[item].notnull()].drop_duplicates(subset=item)[item].sample(n=2))
                ])

dfTitanic_info = pd.DataFrame(data=data, columns=['dataFeatures', 'dataType','null','nullPct','unique','uniqueSample'])
dfTitanic_info
```

### 8. Missing Values
#### Berapa banyak NaN age data dan mean dari setiap who?
#### Isi data NaN dari kolom age dengan mean dari umur tergantung whonya dia apa (child/woman/man)
#### Kemudian print banyak NaN age data dan mean umur setiap who lagi.

```python
nan_man = dfTitanic['age'][dfTitanic['who']=='man'].isna().sum()
avg_age_man = dfTitanic['age'][dfTitanic['who']=='man'].mean()
nan_woman = dfTitanic['age'][dfTitanic['who']=='woman'].isna().sum()
avg_age_woman = dfTitanic['age'][dfTitanic['who']=='woman'].mean()
nan_child = dfTitanic['age'][dfTitanic['who']=='child'].isna().sum()
avg_age_child = dfTitanic['age'][dfTitanic['who']=='child'].mean()
print('Banyak NaN man & mean: ',nan_man,' & ',avg_age_man,
      '\nBanyak NaN woman & mean: ',nan_woman,' & ',avg_age_woman,
      '\nBanyak NaN child & mean: ',nan_child,' & ',avg_age_child)
```

```python
def fill_nan_age(x):
    who = x['who']
    age = x['age']
    mean = dfTitanic.groupby('who')['age'].mean()
    if pd.isnull(age):
        age_n = round(mean[who])
        return age_n
    return age
dfTitanic['age'] = dfTitanic[['who','age']].apply(fill_nan_age, axis=1)
    
```

```python
gb_who = dfTitanic.groupby('who')
for item in dfTitanic['who'].unique():
    mean = gb_who['age'].mean()[item]
    nan = dfTitanic['age'][dfTitanic['who']==item].isna().sum()
    print('Banyak Nan ',item,' & mean:',nan,' & ',mean)
    
```

#### Tambahkan column outliercalcId (FK) di dfTitanic

```python
def out_id(x):
    class_ = x['class']
    if class_ == 'First':
        return 1
    elif class_ == 'Second':
        return 2
    elif class_ == 'Third':
        return 3

dfTitanic['outliercalcId']= dfTitanic.apply(out_id,axis =1)
```

#### Tambahkan column outlier (1 for yes and 0 for no)

```python
def outornot(x):
    class_ = x['class']
    fare = x['fare']
    if class_ == 'First':
        if (fare<df_minmax['batas_max_fare'][0]) & (fare>df_minmax['batas_min_fare'][0]):
            return 0
        else:
            return 1
        
    elif class_ == 'Second':
        if (fare<df_minmax['batas_max_fare'][1]) & (fare>df_minmax['batas_min_fare'][1]):
            return 0
        else:
            return 1
    elif class_ == 'Third':
        if (fare<df_minmax['batas_max_fare'][2]) & (fare>df_minmax['batas_min_fare'][2]):
            return 0
        else:
            return 1


dfTitanic['outlier']= dfTitanic.apply(outornot,axis =1)
```
